# Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
```

```python
num_col=['BP','Cholestral','Max HR','Age']
sc=StandardScaler()
df[num_col]=sc.fit_transform(df[num_col])
df
```

#### Splitting the data into train and test

```python
df['Heart Disease'].replace({'Presence':1,'Absence':0},inplace=True)
df
x=df[['Age','Max HR','BP','Cholestral','Chest Pain Type','FBS','EKG','Number of Vessels Fluro','Thallium','Exercise Angina','Slope of ST']]
```

```python
df.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,df['Heart Disease'],test_size=0.2,random_state=42)
```

#### Logistic Regression

```python
logistic=LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
y_pred_logistic=logistic.predict(x_test)
acc=accuracy_score(y_test,y_pred_logistic)
print("Accuracy of logistic regression is",acc)
print("\nclassificatiopn report\n",classification_report(y_test,y_pred_logistic))
```

#### Random Forest

```python
random=RandomForestClassifier()
random.fit(x_train,y_train)
```

```python
y_pred_random=random.predict(x_test)
acc=accuracy_score(y_test,y_pred_random)
print("Accuracy of Random Forest is",acc)
print("\nclassificatiopn report\n",classification_report(y_test,y_pred_random))
```

#### Decision Tree

```python
decision=DecisionTreeClassifier()
decision.fit(x_train,y_train)
```

```python
y_pred_decision=decision.predict(x_test)
acc=accuracy_score(y_test,y_pred_decision)
print("Accuracy of Decision Tree is",acc)
print("\nClassification report\n",classification_report(y_test,y_pred_decision))
```

##### K-Nearset neighbour

```python
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
```

```python
y_pred_KNN=KNN.predict(x_test)
acc=accuracy_score(y_test,y_pred_KNN)
print("Accuracy of KNN is",acc)
print("\nClassification report\n",classification_report(y_test,y_pred_KNN))
```

##### Support Vector Machine

```python
svm=SVC(kernel='linear')
svm.fit(x_train,y_train)
```

```python
y_pred_svm=svm.predict(x_test)
acc=accuracy_score(y_test,y_pred_svm)
print("Accuracy of SVM is",acc)
print("\nClassification report\n",classification_report(y_test,y_pred_svm))
```

#### Naive Bayes

```python
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_pred_nb=nb.predict(x_test)
acc=accuracy_score(y_test,y_pred_nb)
print("Accuracy of Naive Bayes is",acc)
print("\nClassification report\n",classification_report(y_test,y_pred_nb))
```
